```python
from dataclasses import dataclass

import pandas as pd

from fastlite import database
```

```python
df = pd.read_csv("numbers.csv", index_col=0)
```

```python
df.head()
```

```python
# 1) Create the class that defines your table schema
@dataclass
class Numbers:
    number: int
    neoficialiai: str
    compound: str
    years: str
    kokia_kaina: str
    kokia_kaina_compound: str
    euro_nom: str
    cent_nom: str
    kiek_kainuoja: str
    kiek_kainuoja_compound: str
    euro_acc: str
    cent_acc: str

# 2) Create or open the DB
db = database("lithuanian_data.db")

# 3) Create the table if it doesn't exist
tbl_name = "numbers"
if tbl_name not in db.t:
    db.create(Numbers, name=tbl_name, pk="number")
```

```python
db.t[tbl_name].insert_all(df.to_dict(orient='records'))
```

```python
# 6) Query the DB
rows = db.q("SELECT * FROM numbers LIMIT 5;")
print(rows)
```

```python
@dataclass
class Attempt:
    user_id: int
    number: int
    correct: bool
    timestamp: str
    grammar_tested: str  # Adding the new grammar_tested field
    exercise_id: int     # Adding the new exercise_id field

# 4) Create the 'attempts' table
attempts_tbl = "attempts"
if attempts_tbl not in db.t:
    db.create(Attempt, name=attempts_tbl)

# 5) Insert an example attempt
db.t[attempts_tbl].insert({
    "user_id": 101,
    "number": 1,
    "correct": True,
    "timestamp": "2025-02-16 12:34:56",
    "grammar_tested": "accusative",  # Example grammar type being tested
    "exercise_id": 1                 # Example exercise ID
})
```

```python
schema_query = """
SELECT name, sql 
FROM sqlite_master 
WHERE type='table';
"""
print("\nDatabase Schema:")
for row in db.query(schema_query):
    print(f"\n{row}")
```
